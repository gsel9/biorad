```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
results_sbs = pd.read_csv(
    './../../../data_source/results/dtree_univariate_sbs/50evals_50reps_dtree_univariate_sbs_dfs.csv',
    index_col=0
)
results_sbs.head()
```

```python
results_sfs = pd.read_csv(
    './../../../data_source/results/dtree_univariate_sfs/50evals_50reps_dtree_univariate_sfs_dfs.csv',
    index_col=0
)
results_sfs.head()
```

```python
sfs_model_grps = results_sfs.groupby('experiment_id')
_sfs_avg_scores = sfs_model_grps['test_score'].mean()
_sfs_rows, _sfs_cols = zip(*list(map(lambda label: label.split('_'), _sfs_avg_scores.index)))
sfs_rows, sfs_cols = np.unique(_sfs_rows), np.unique(_sfs_cols)
sfs_avg_scores = sfs_model_grps['test_score'].mean().values.reshape(np.size(sfs_cols), np.size(sfs_rows))
sfs_std_scores = sfs_model_grps['test_score'].std().values.reshape(np.size(sfs_cols), np.size(sfs_rows))
```

```python
sbs_model_grps = results_sbs.groupby('experiment_id')
_sbs_avg_scores = sbs_model_grps['test_score'].mean()
_sbs_rows, _sbs_cols = zip(*list(map(lambda label: label.split('_'), _sbs_avg_scores.index)))
sbs_rows, sbs_cols = np.unique(_sbs_rows), np.unique(_sbs_cols)
sbs_avg_scores = sbs_model_grps['test_score'].mean().values.reshape(np.size(sbs_cols), np.size(sbs_rows))
sbs_std_scores = sbs_model_grps['test_score'].std().values.reshape(np.size(sbs_cols), np.size(sbs_rows))
```

```python
sfs_model_name = 'Sequential Forward Floating Selection\n&'
sbs_model_name = 'Sequential Backward Floating Selection\n&'

selector_names = [
    'ANOVA F-value', r'$\chi^2$-test', 'Fisher Score',
    'Minimum Redundancy \nMaximum Relevancy', 
    'Mutual Information', 'ReliefF', 
    'Student t-test',
    'Wilcoxon rank-sum'
]
fig, (sfs_axis, cbar_axis, sbs_axis) = plt.subplots(
    nrows=3, 
    gridspec_kw={'height_ratios': (1.5, 0.05, 1.3), 'hspace': -1.07}, 
    figsize=(8, 8)
)
fig.subplots_adjust(wspace=0.0001)
sns.heatmap(
    sfs_avg_scores,
    yticklabels='',
    vmin=np.nanmin(sfs_avg_scores) - 1e-4, 
    vmax=np.nanmax(sfs_avg_scores) + 1e-4,
    cmap=plt.cm.viridis, 
    robust=True,
    annot=True, 
    fmt='.3f', 
    square=1, 
    linewidth=.2, 
    ax=sfs_axis,
    cbar=True,
    cbar_ax=cbar_axis,
    cbar_kws={'shrink': 0.2, 'orientation': 'horizontal'},  
    
)
sns.heatmap(
    sbs_avg_scores,
    yticklabels='',
    vmin=np.nanmin(sbs_avg_scores) - 1e-4, 
    vmax=np.nanmax(sbs_avg_scores) + 1e-4,
    cmap=plt.cm.viridis, 
    robust=True,
    annot=True, 
    fmt='.3f', 
    square=1, 
    linewidth=.2, 
    ax=sbs_axis,
    cbar=False,
    cbar_kws={'shrink': 0.2, 'orientation': 'horizontal'},  
    
)
sfs_axis.set_xticklabels(selector_names, ha='center')
sbs_axis.set_xticklabels(selector_names, ha='center')
# Position the text.
sfs_axis.text(
    0.5, 1.4, 
    sfs_model_name, 
    transform=sbs_axis.transAxes, 
    fontsize=14,    
    va='center',
    ha='center'
)
sfs_axis.tick_params(rotation=45)
sbs_axis.text(
    0.5, 1.4, 
    sbs_model_name, 
    transform=sfs_axis.transAxes, 
    fontsize=14,    
    va='center',
    ha='center'
)
sbs_axis.tick_params(rotation=45)
plt.savefig(
    './../../../figures/50vals_50reps_dtree_univariate_sbs_dfs.png',
    bbox_inches='tight',
    transparent=True,
    dpi=100, 
)
```
